```python
import os
import numpy as np
import natsort as n

from sklearn import svm

import matplotlib.pyplot as plt
```

```python
root = '/home/pouple/PhD/Code/Dimmy/Stims'

PT_paths = [os.path.join(root, f) for f in os.listdir(root) if 'PT' in f]
PTs = [np.load(f) for f in n.natsorted(PT_paths)]
names = [os.path.basename(f)[:-4] for f in n.natsorted(PT_paths)]
```

```python
clf = svm.SVC()
X = [PTs[0], PTs[-1]]
clf.fit(X, [1, 0])
y_pred = clf.predict(PTs)
plt.plot(y_pred)
```
